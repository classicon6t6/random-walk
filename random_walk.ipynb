```python
import numpy as np
```

```python
#Climbing stairs with rolled die, step -1 when rolled 1 or 2, step +0 when rolled 3, +1 when rolled 4 or 5 and when rolled 6 step + next rolled die

#list of end level for 1000 trials
all_level = []
```

```python
for y in range(1000) :
	#starting level
	level = [0]
	for x in range(100) :
		step = level[-1]

		#roll the die
		die = np.random.randint(1,7)

		#stepping condition
		if die <= 2:
			step = max(0, step - 1)
		elif die == 3 :
			step = step
		elif die <= 5 :
			step = step + 1
		else :
			step = step + np.random.randint(1,7)
	
	#save in level
		level.append(step)
	all_level.append(level)
#rolling the die for 100 times
#randint : generat random integer from a to b-1 for a,b in Z
```

```python
#this shows how high are you
print(all_level)
```

```python
#required package for visualizing data
import matplotlib.pyplot as plt
```

```python
#plotting level
graph = plt.plot(level)
plt.xlabel('Steps')
plt.ylabel('Level')
plt.title('Last Run\'s Outcome')

#making arrow
plt.annotate('Highest value\nin last run'+'\n= ' + str(max(level)), xy=(level.index(max(level)), max(level)), xytext=(level.index(max(level))-45, max(level)-15),
arrowprops=dict(facecolor='black', shrink=0.05),)

#saving the plot in png file
plt.savefig('Last Run.png')
```

```python

#clearing plot so it doesn't appear in next graph
plt.clf()
#making histogram for outome's distribution
#convert all_level to np array
np_all = np.transpose(np.array(all_level))
end_level = np_all[-1][:]
print(end_level)
```

```python
mu = end_level.mean()
sigma = end_level.std()

#ploting histogram
plt.hist(end_level, bins = 20, facecolor = 'grey')
plt.xlabel('Level')
plt.ylabel('Sum of Level')
plt.title('Distribution of All Outcome')

plt.text(20, 125, r'$\mu=%.2f$' % (mu, ))
plt.text(20, 120, r'$\sigma=%.2f$' % (sigma, ))

plt.grid(True)
plt.savefig('Distribution of Outcome.png')
```
